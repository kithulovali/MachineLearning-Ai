```python
import numpy as np
from numpy import random
# randomise array1
arr1 = random.randint(100 ,size=(12))
print(arr1)
# randomise array 2 
arr2 = random.randint(10 , size=(12))
print(arr2)
# array addition 
arr3 = np.add(arr1,arr2)
# array subtraction
arr4 = np.subtract(arr1,arr2)
# array multiplication
arr5 = np.multiply(arr1,arr2)
# array division
arr6 = np.divide(arr1,arr2)
print("devision :",arr6)
print("multiplication :",arr5)
print("subtruction :",arr4)
print("sum :",arr3)
```
