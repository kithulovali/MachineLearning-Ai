```python
import matplotlib.pyplot as plt
import numpy as np 
from numpy import random
# definning the axis 
x_axis = np.array([1,8])
y_axis = np.array([1,10])

#plotting the graph 
plt.plot(x_axis,y_axis)
plt.show()
```

```python
# ploting mutiple ponts 
x_axis = np.array([1,2,3,4])
y_axis = np.array([3,8,1,10])

plt.plot(x_axis , y_axis)
plt.show()
```

```python
y_axis = random.randint(5 ,size=5)
print(y_axis)
plt.plot(y_axis ,marker = 'o')
plt.show()
```

```python
y_axis = random.randint(5, size =5)
plt.plot(y_axis , c='g', ls ='dotted' ,marker ='o')
plt.show()
```

```python
#x_axis = random.randint(5 , size = 5)
y_axis = random.randint(5, size = 5)
plt.plot(y_axis , marker = '*', c = 'r' ,ls= 'dotted') 
plt.xlabel('random Xs')
plt.ylabel('random Ys') 
plt.grid()
plt.show() 
```
