```python
#x_axis = random.randint(5 , size = 5)
y_axis = random.randint(5, size = 5)
plt.plot(y_axis , marker = '*', c = 'r' ,ls= 'dotted') 
plt.xlabel('random Xs')
plt.ylabel('random Ys') 
plt.grid()
plt.show() 
```

```python
# multiples plots on one figure 
# figure one 
#x_axis = random.randint(5 , size = 5)
y_axis = random.randint(5 , size = 5)
plt.subplot(2,1,1)
plt.plot(y_axis , marker ='*', c ='g')
plt.grid()
plt.xlabel('random Xs figure 1')
plt.ylabel('random Ys figure 1')
plt.title('ramdon figure 1')
plt.show()
#x_axis = random.randint(5 , size = 5)
y_axis = random.randint(5 , size = 5)
plt.subplot(2,1,2)
plt.grid()
plt.plot(y_axis , marker ='o',c= 'r')
plt.xlabel('random Xs figure 2')
plt.ylabel('random Ys figure 2')
plt.title('random figure 2 ')
plt.show()
```
