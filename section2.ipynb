```python

import numpy as np 
from numpy import random
```

```python
"""
Exercise 6

Create the following NumPy array:
[[1, 2, 3],
 [4, 5, 6]]

Then print:
The shape of the array
The data type of the array
The number of dimensions
"""
arr = np.array([[1,2,3],[4,5,6]])
print(arr)
print(arr.ndim) # printing dimensions
print(arr.shape) # printing the shape
print(arr.dtype) # printing the datatype
```

```python
from numpy import random
"""
Create a one-dimensional NumPy array containing 12 elements.
Then:
Reshape the array into a 3Ã—4 matrix
Print the reshaped array
Print the new shape of the array
"""
arr = random.randint(100, size=(12))
print(arr.ndim)
Rarr = arr.reshape(3,4)
print(arr)
print(Rarr)
print(Rarr.shape)
```
