```python
# Task one 
import pandas as pd 
# loading csv 
students_df = pd.read_csv('student_performance.csv')
# converting csv into DataFrame
student_data = pd.DataFrame(students_df) 
print(student_data)
# getting top 5 elements 
print(student_data.head(5))
# checking datatypes 
print(student_data.info())
# counting  missing values per column 
print(student_data.isna().sum())
```

```python
# task two 
# finging  empty cells in columns
mean_math = student_data['math'].mean()
mean_english = student_data['english'].mean()
mean_science = student_data['science'].mean()
# filling emty cells with the mean values
student_data['math'] = student_data['math'].fillna(mean_math)
student_data['english'] = student_data['english'].fillna(mean_english)
student_data['science'] = student_data['science'].fillna(mean_science)
print(student_data)
# checking for empty cells 
print(student_data.isna())
```

```python
# task 3 adding average row and pass row
# average column
student_data['average'] = student_data['math'] + student_data['english'] + student_data['science'] /3
print(student_data)
# passed column
student_data['passed'] = student_data['average'] >= 60
print(student_data)
```

```python
# creating column attendence_ok
student_data['attendance_ok'] = student_data['attendance'] >= 75
print(student_data)
```

```python
# grading student  according to existing data 
import numpy as np
conditions = [
    student_data['average'] >= 80,
    student_data['average'] >= 70,
    student_data['average'] >= 60
]
choises = ['A','B','C']
student_data['grades'] = np.select(conditions , choises , default ='F')
print(student_data)
```

```python
# eligibility for exams 
student_data['exam_pass'] = student_data['passed'] & student_data['attendance_ok']
print(student_data)
```
