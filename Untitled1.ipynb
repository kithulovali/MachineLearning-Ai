```python
import matplotlib.pyplot as plt
import numpy as np 
from numpy import random
# definning the axis 
x_axis = np.array([1,8])
y_axis = np.array([1,10])

#plotting the graph 
plt.plot(x_axis,y_axis)
plt.show()
```

```python
# ploting mutiple ponts 
x_axis = np.array([1,2,3,4])
y_axis = np.array([3,8,1,10])

plt.plot(x_axis , y_axis)
plt.show()
```
